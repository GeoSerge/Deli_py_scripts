```python
import pandas as pd
import numpy as np
import re
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('C:/Users/sgulbin/Work/Analysis/FTR_score_v2/test_roc_auc_v2.csv')
df['1'] = 1
df['K_final_weighted_inverted'] = df['1']-df['K_final_sq']
# Running roc-auc score calculation
roc_auc_score(df['accident_probability'],df.K_final_weighted_inverted)
```

```python
# for i in np.linspace(0.01,1,100):
# result_df = pd.DataFrame({'sex_w':[], 'age_exp_w':[], 'kbm_w':[], 'roc-auc':[]})
all_sex_w = []
all_age_exp_w = []
all_kbm_w = []
all_lcns_w = []
all_psp_w = []
roc_auc_scores = []
sex_w = [i/100 for i in range(0,11,1)] #np.linspace(0.01,1,100)
age_exp_w =  [i/100 for i in range(40,51,1)] #np.linspace(0.01,1,100)
kbm_w =  [i/100 for i in range(40,51,1)] #np.linspace(0.01,1,100)
license_w =[i/100 for i in range(0,11,1)]#[i/100 for i in range(0,105,5)] #np.linspace(0.01,1,100)
psp_code_w =  [i/100 for i in range(5,21,1)] #np.linspace(0.01,1,100)
for s_index,sex in enumerate(sex_w):
    for ae_index,ae in enumerate(age_exp_w):
        print(s_index,ae_index)
        for kbm in kbm_w:
            for lcns in license_w:
                for psp in psp_code_w:
                    if sex+ae+kbm+lcns+psp == 1.0:
#                         print('running')
                        loop_df = df
                        loop_df['1'] = 1
                        loop_df['sex_w'] = sex
                        loop_df['age_exp_w'] = ae
                        loop_df['kbm_w'] = kbm
                        loop_df['lcns_w'] = lcns
                        loop_df['psp_code_w'] = psp
                        loop_df['K'] = sex*loop_df.K_sex**2+ae*loop_df['K_age-exp']**2+kbm*loop_df.K_kbm**2+lcns*loop_df.K_license**2+psp*loop_df.K_bp_region**2
                        loop_df['K'] = df['1']-df['K']
                        all_sex_w.append(sex)
                        all_age_exp_w.append(ae)
                        all_kbm_w.append(kbm)
                        all_lcns_w.append(lcns)
                        all_psp_w.append(psp)
                        roc_auc_scores.append(roc_auc_score(df['accidents_fact'],df.K))
output_df = pd.DataFrame({'sex_weight':all_sex_w, 'age_exp_weight':all_age_exp_w, 'kbm_weight': all_kbm_w, 'license_weight':\
                          all_lcns_w, 'psp_district_weight':all_psp_w, 'roc_auc':roc_auc_scores})
output_df['total_weight'] = output_df.sex_weight+output_df.age_exp_weight+output_df.kbm_weight+output_df.license_weight+output_df.psp_district_weight
print(output_df)
```
