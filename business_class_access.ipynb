```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/sgulbin/Work/Analysis/BusinessClassAccess/data.csv')
# Subtracting taxes from cost and order sum
df['cost'] = df['cost']/1.2
df['order_sum'] = -df['order_sum']/1.2
df['ue_cost'] = np.where(df.vehicle_brand == 'BMW', df.rent_time*-7.66, df.rent_time*-6.06)
output = pd.DataFrame({'score':[],'rent_time_AS_IS':[],'rent_time_TO_BE':[],'ride_time_AS_IS':[],'ride_time_TO_BE':[],\
                       'cost_AS_IS':[], 'cost_TO_BE':[],'order_sum_AS_IS':[],'order_sum_TO_BE':[], 'ue_cost_AS_IS':[],\
                       'ue_cost_TO_BE':[],'econ_effect_AS_IS':[],'econ_effect_TO_BE':[],'econ_effect_UE_AS_IS':[],\
                       'econ_effect_UE_TO_BE':[]})
for score in np.arange(0.01,1.01,0.01):
    dfr = df
    dfr['cost_TO_BE'] = np.where(dfr.final_score < score, 0, dfr.cost)
    dfr['order_sum_TO_BE'] = np.where(dfr.final_score < score, 0, dfr.order_sum)
    dfr['ride_time_TO_BE'] = np.where(dfr.final_score < score, 0, dfr.ride_time)
    dfr['rent_time_TO_BE'] = np.where(dfr.final_score < score, 0, dfr.rent_time)
    dfr['ue_cost_TO_BE'] = np.where(dfr.final_score < score, 0, dfr.ue_cost)
    result = pd.DataFrame({'score':[score]\
                           ,'rent_time_AS_IS': [dfr.rent_time.sum()]\
                           ,'rent_time_TO_BE': [dfr.rent_time_TO_BE.sum()]\
                           ,'ride_time_AS_IS': [dfr.ride_time.sum()]\
                           ,'ride_time_TO_BE': [dfr.ride_time_TO_BE.sum()]\
                           ,'cost_AS_IS': [dfr.cost.sum()]\
                           ,'cost_TO_BE': [dfr.cost_TO_BE.sum()]\
                           ,'order_sum_AS_IS': [dfr.order_sum.sum()]\
                           ,'order_sum_TO_BE': [dfr.order_sum_TO_BE.sum()]\
                           ,'ue_cost_AS_IS': [dfr.ue_cost.sum()]\
                           ,'ue_cost_TO_BE': [dfr.ue_cost_TO_BE.sum()]\
                           ,'econ_effect_AS_IS':[dfr.cost.sum()+dfr.order_sum.sum()]\
                           ,'econ_effect_TO_BE':[dfr.cost_TO_BE.sum()+dfr.order_sum_TO_BE.sum()]\
                           ,'econ_effect_UE_AS_IS':[dfr.cost.sum()+dfr.order_sum.sum()+dfr.ue_cost.sum()]\
                           ,'econ_effect_UE_TO_BE':[dfr.cost_TO_BE.sum()+dfr.order_sum_TO_BE.sum()+dfr.ue_cost_TO_BE.sum()]})
    output = output.append(result)
output.to_csv('C:/Users/sgulbin/Work/Analysis/BusinessClassAccess/econ_effect_v2.csv', index = False)
    
    
```
