```python
from dadata import Dadata
import pandas as pd
import numpy as np
```

```python
def dadata(df: pd.DataFrame, token: str, secret: str):
    end2 = {}
    for num, i in enumerate(df.birth_place.fillna('')):
        if not i:
            end2[num] = np.nan
            continue
        dadata = Dadata(token, secret)
        result = dadata.suggest("address", i)
        end2[num] = result
        if num % 100 == 0:
            print(num)
    temp = []
    for i in range(len(end2)):
        try:
            temp.append(end2[i][0]['data']['region_with_type'])
        except (TypeError, IndexError):
            temp.append(end2[i])

    for i in range(len(temp)):
        if isinstance(temp[i], list):
            temp[i] = np.nan 
    df['bp_loc_name'] = temp
    return df
```
