```python
import pandas as pd
import numpy as np

# Upload repairs data
dtp = pd.read_excel('C:/Users/sgulbin/Work/Data/accidents/DTP_data_7_16_2020.xlsx')
dtp['Дата и время ДТП'] = pd.to_datetime(dtp['Дата и время ДТП'], dayfirst=False, errors = 'coerce')
dtp['yyyy-mm'] = dtp['Дата и время ДТП'].dt.strftime('%Y-%m')
dtp['date'] = dtp['Дата и время ДТП'].dt.date
dtp['time'] = dtp['Дата и время ДТП'].dt.time
dtp['weekday'] = dtp['Дата и время ДТП'].dt.day_name()
dtp['day_of_year'] = dtp['Дата и время ДТП'].dt.dayofyear
dtp.sort_values('Дата и время ДТП', inplace=True)
dtp['Телефон водителя'] = dtp['Телефон водителя'].replace(np.nan, 0, regex=True)
dtp['Телефон водителя'] = dtp['Телефон водителя'].apply(int)
dtp['Сумма по заказ-наряду'] = dtp['Сумма по заказ-наряду'].apply(float)
dtp['Выставлено'] = dtp['Выставлено'].apply(float)
dtp['Страховая компания сумма выплаченная'] = dtp['Страховая компания сумма выплаченная'].apply(float)
# dtp.reset_index(drop=True, inplace = True)
dtp['ТС'] = dtp['ТС'].str.replace(' ', '')
dtp = dtp[['IDИнциндента','Дата и время ДТП','yyyy-mm','date','time','weekday','day_of_year','Регион','Телефон водителя','Виновность', \
           'ТС', 'Модель автомобиля', 'POOL', 'Тариф','Ремонт не производился','Тип ДТП', 'Вид страхования', 'Пострадавшие',\
           'Комментарий','Сумма по заказ-наряду','Выставлено', 'Страховая компания сумма выплаченная', 'Виновник пьян',\
           'Водитель находился в состояние опьянения','Клиент скрылся','Тотал','Машина в розыске','Наличие опьянения',\
           'Номер ОСАГО','Номер убытка']].rename(columns={
    'IDИнциндента': 'ticket_id',
    'Дата и время ДТП': 'datetime',
    'yyyy-mm': 'yyyy-mm',
    'date':'date',
    'time':'time',
    'weekday': 'weekday',
    'day_of_year':'day_of_year',
    'Регион': 'region',
    'Телефон водителя': 'phone',
    'Виновность': 'responsibility',
    'ТС': 'license_plate',
    'Модель автомобиля': 'vehicle_model',
    'POOL': 'pool',
    'Тариф': 'tariff',
    'Ремонт не производился': 'no_repair_conducted',
    'Тип ДТП': 'accident_type',
    'Вид страхования': 'insurance_type',
    'Пострадавшие': 'injured',
    'Комментарий': 'description',
    'Сумма по заказ-наряду': 'repair_cost',
    'Выставлено': 'vystavleno',
    'Страховая компания сумма выплаченная': 'insurance_paid_sum',
    'Виновник пьян':'accident_causer_drunk',
    'Водитель находился в состояние опьянения': 'driver_was_drunk',
    'Клиент скрылся':'run_driver',
    'Тотал':'salvage',
    'Машина в розыске':'vehicle_wanted',
    'Наличие опьянения':'DUI',
    'Номер ОСАГО':'OSAGO_policy_number',
    'Номер убытка':'loss_id' 
})
dtp.to_csv('C:/Users/sgulbin/Work/Data/accidents/accidents_data.csv', index = False)
# dtp.to_excel('C:/Users/sgulbin/Work/Data/accidents/accidents_data.xlsx', index = False)
```

```python
dtp['dt_y-m'] = pd.to_datetime(dtp['dt']).dt.to_period('M')
dtp['repair_data_availability'] = np.where(dtp.target_amount > 0,1,0)
total_repairs = dtp.groupby('dt_y-m').count()
total_repairs = total_repairs[['dt']]
nonempty_repairs = dtp.groupby('dt_y-m').sum()
nonempty_repairs = nonempty_repairs[['repair_data_availability']]
data_qual = pd.merge(total_repairs, nonempty_repairs, left_index = True, right_index = True)
data_qual['repair_data_availability, %'] = data_qual['repair_data_availability']/data_qual['dt']
data_qual
# data_qual.to_csv('C:/Users/sgulbin/Work/Data/accidents/accidents_repair_sum_availability.csv')
```
