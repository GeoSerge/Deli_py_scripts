```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/sgulbin/Work/_Ad-hoc_analysis/usrs_card_nmbrs.csv', low_memory=False)
active = pd.read_csv('C:/Users/sgulbin/Work/_Ad-hoc_analysis/usrs_drove_after_6_9_2020.csv')
active['active'] = 1
df['card_number'] = pd.to_numeric(df['card_number'], errors = 'coerce')
df['card_number'] = df['card_number'].fillna(0)
df['card_number'] = df['card_number'].astype(int)
bins = pd.read_csv('C:/Users/sgulbin/Work/_Ad-hoc_analysis/bins.csv')
bins = bins[['6digits','bank']]
df = pd.merge(df,bins, left_on = 'card_number', right_on = '6digits', how = 'left')
df = pd.merge(df, active, on = 'user_id', how = 'left')
df_active = df[df['active'] == 1]
df
```
