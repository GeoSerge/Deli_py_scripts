```python
import pandas as pd
import numpy as np
import re
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/korischenko_validation/Oct-Nov_validation.csv')
dtp = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/economic_effect_v2/DTP_data.csv')
users_ext = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/korischenko_validation/dtp_users_ext.csv')

# Adding user_ext to dtp data
dtp = pd.merge(dtp, users_ext, left_on = 'user_id', right_on = 'user_id', how = 'left')
dtp['user_ext'] = dtp.user_ext.fillna(0)
dtp.user_ext = dtp.user_ext.astype(int)
dtp['date'] = pd.to_datetime(dtp.date)
# Splitting dtp data into periods for validation
# October-November
dtp_oct_nov = dtp.loc[(dtp.date >= '2019-10-01') & (dtp.date <= '2019-11-30')]
dtp_oct_nov = dtp_oct_nov[['date','user_ext']]
dtp_oct_nov['oct_nov'] = 1
# Marking users with accidents
df = pd.merge(df,dtp_oct_nov, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df['oct_nov'] = df['oct_nov'].fillna(0)
```

```python
for i in range(2,11):
    df1 = df.loc[df['rents_count']>= i]
    print(i,roc_auc_score(df1['oct_nov'],df1.probability))
```

```python
df = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/korischenko_validation/Oct-Nov_validation.csv')
dtp = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/economic_effect_v2/DTP_data.csv')
users_ext = pd.read_csv('C:/Users/sgulbin/Work/Analysis/ScoringTesting/korischenko_validation/dtp_users_ext.csv')

# Adding user_ext to dtp data
dtp = pd.merge(dtp, users_ext, left_on = 'user_id', right_on = 'user_id', how = 'left')
dtp['user_ext'] = dtp.user_ext.fillna(0)
dtp.user_ext = dtp.user_ext.astype(int)
dtp['date'] = pd.to_datetime(dtp.date)
# Splitting dtp data into periods for validation
# 1st week of December
dtp_1st_w = dtp.loc[(dtp.date >= '2019-12-01') & (dtp.date <= '2019-12-07')]
dtp_1st_w = dtp_1st_w[['date','user_ext']]
dtp_1st_w['1st_w'] = 1
# December
dtp_dec = dtp.loc[(dtp.date >= '2019-12-01') & (dtp.date <= '2019-12-31')]
dtp_dec = dtp_dec[['date','user_ext']]
dtp_dec['Dec'] = 1
# October-November
dtp_oct_nov = dtp.loc[(dtp.date >= '2019-10-01') & (dtp.date <= '2019-11-30')]
dtp_oct_nov = dtp_oct_nov[['date','user_ext']]
dtp_oct_nov['oct_nov'] = 1
# 1st week of October
dtp_oct1_7 = dtp.loc[(dtp.date >= '2019-10-01') & (dtp.date <= '2019-10-07')]
dtp_oct1_7 = dtp_oct1_7[['date','user_ext']]
dtp_oct1_7['oct1_7'] = 1
# October
dtp_oct = dtp.loc[(dtp.date >= '2019-10-01') & (dtp.date <= '2019-10-31')]
dtp_oct = dtp_oct[['date','user_ext']]
dtp_oct['oct'] = 1
# November
dtp_nov = dtp.loc[(dtp.date >= '2019-11-01') & (dtp.date <= '2019-11-30')]
dtp_nov = dtp_nov[['date','user_ext']]
dtp_nov['nov'] = 1
# Marking users with accidents
df = pd.merge(df,dtp_1st_w, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df = pd.merge(df,dtp_dec, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df = pd.merge(df,dtp_oct_nov, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df = pd.merge(df,dtp_oct, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df = pd.merge(df,dtp_nov, left_on = 'user_ext', right_on = 'user_ext', how = 'left')
df = pd.merge(df,dtp_oct1_7, left_on = 'user_ext', right_on = 'user_ext', how = 'left')

df['oct_nov'] = df['oct_nov'].fillna(0)
df['1st_w'] = df['1st_w'].fillna(0)
df['Dec'] = df['Dec'].fillna(0)
df['oct'] = df['oct'].fillna(0)
df['nov'] = df['nov'].fillna(0)
df['oct1_7'] = df['oct1_7'].fillna(0)
```

```python
for period in ['oct1_7','oct','nov','oct_nov','1st_w','Dec']:
    print(period)
    for i in range(2,11):
        df1 = df.loc[df['rents_count']>= i]
        print(roc_auc_score(df1[period],df1.probability))
```
