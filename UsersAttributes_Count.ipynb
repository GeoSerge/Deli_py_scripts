```python
## RUNNING SQL QUERY WITH VERTICA_PYTHON
import pandas as pd
import vertica_python
conn_info = {'host': '192.168.246.202',
             'port': 5433,
             'user': 'sgulbin',
             'password': 'SG.300390',
             'database': 'delimobil_dwh'}

fd = open('C:/Users/sgulbin/Work/Scripts (Queries)/ClientsData.sql', 'r')
sqlFile = fd.read()
fd.close()

with vertica_python.connect(**conn_info) as conn:
    cur = conn.cursor()
    cur.execute(sqlFile)
    df = pd.DataFrame(cur.fetchall(), columns = [desc[0] for desc in cur.description ])
#print(df)
```
