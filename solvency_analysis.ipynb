```python
import matplotlib.pyplot as plt
def solvency_calc(target_df, target_field, total_inv_column, rent_inv_column, camera_inv_column, accidents_inv_column, \
                  agreement_inv_column, other_inv_column,total_invoices, total_paid, rent_invoices, rent_paid, camera_invoices,\
                  camera_paid, accidents_invoices,accidents_paid,agreement_invoices,agreement_paid, other_invoices, other_paid):
    dfr = target_df.groupby(target_field).sum()
#     dfr = dfr.loc[(dfr.index > 18)&(dfr.index <51)]
    dfr[total_inv_column] = dfr[total_paid]/dfr[total_invoices]
    dfr[rent_inv_column] = dfr[rent_paid]/dfr[rent_invoices]
    dfr[camera_inv_column] = dfr[camera_paid]/dfr[camera_invoices]
    dfr[accidents_inv_column] = dfr[accidents_paid]/dfr[accidents_invoices]
    dfr[agreement_inv_column] = dfr[agreement_paid]/dfr[agreement_invoices]
    dfr[other_inv_column] = dfr[other_paid]/dfr[other_invoices]
    dfr = dfr[[total_inv_column,rent_inv_column,camera_inv_column,accidents_inv_column,agreement_inv_column,other_inv_column]]
    dfr.plot()
    print(dfr)

solvency_calc(df, 'bill_total', 'total_%_paid', 'rent_%_paid', 'camera_%_paid','accident_%_paid','agreement_%_paid','other_%_paid',\
              'total_invoices_sum','total_paid_invoices_sum','rent_invoices_sum','paid_rent_invoices_sum','camera_invoices_sum',\
              'paid_camera_invoices_sum', 'accident_invoices_sum', 'paid_accident_invoices_sum','agreement_invoices_sum',\
              'paid_agreement_invoices_sum', 'other_invoices_sum', 'paid_other_invoices_sum')
```
